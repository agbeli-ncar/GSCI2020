## Regression
For the first example of machine learning, we will explore different types of regression. Many are familar with a scatter plot and a line-of-best fit. However this is a form of machine learning because the the line-of-best-fit can be used to predict intermediate values. 




```python
# importing modules 
import numpy as np
import matplotlib.pyplot as plt
import pandas
```

```python
# Creating numpy arrays
number_of_values = 20
#values=np.random.randint(100, size=(number_of_values))
#x = np.arange(1, number_of_values+1, 1)

df = pandas.DataFrame(np.random.randint(0,100,size=(100, 2)), columns=['A','Values'])

print(df)
```

```python
df.plot.scatter(x='A', y='Values');
```

```python
# Creating a function named regression that calculates the slope and intercept of 
# the line-of-best-fit

def regression(x,y):
    """ This function calculates a simple linear regression
    This is done by checking that the inputs are 1-D
    Arrays and then find the slope and intercept. """
    a = (sum(y)*sum(np.square(x)) - sum(x)*sum(x*y))/\
        (len(x)*sum(np.square(x))-sum(x)*sum(x))

    b = (len(x)*sum(x*y) - sum(x)*sum(y))/\
        (len(x)*sum(np.square(x))-(sum(x)*sum(x)))
    
    #print(a)
    #print(b)

    return a,b
```

```python
[a,c] = regression(df['A'],df['Values'])
print(a)
print(c)

fig,ax = plt.subplots()
ax.scatter(df['A'],df['Values'])
ax.plot(df['A'],a+c*df['A'],color='r')
ax.set_xlabel('Indexes')
ax.set_ylabel('Values')
ax.set_xticks([1,5,10,15,20])

ax.text(.9, .9, 
        'r = {} '.format(round(c,2)), 
        horizontalalignment='center',
        verticalalignment='center', 
        transform=ax.transAxes,
        bbox=dict(facecolor='none', 
                  edgecolor='black', 
                  boxstyle='square'))


#df.plot.scatter(x='A', y='Values');
#ax.plot(df['A'],a+c*df['A'],color='r')
```
